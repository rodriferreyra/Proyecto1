```python
# Reemplazamos los valores nulos de 3 columnas, se crea una variable nueva

valores_reemplazo = {
    'director': 'Director no especificado',
    'cast': 'Elenco no disponible',
    'country': 'País no disponible'
}

Netflix.fillna(valores_reemplazo, inplace=True)

print(Netflix.isnull().sum())

```

```python
# Cambiamos el formato de fecha en la columna date_added

# Aseguramos de que la columna sea de tipo string

Netflix['date_added'] = Netflix['date_added'].astype(str)

Netflix['date_added'] = pd.to_datetime(Netflix['date_added'], format='%B %d, %Y', errors='coerce')

# Queremos el formato dd/mm/aaaa como cadena

Netflix['date_added'] = Netflix['date_added'].dt.strftime('%d/%m/%Y')

# Imprimmimos y vemos las primera 5 fechas

print(Netflix['date_added'].head())
```

```python
# Creamos 2 columnas nuevas, vamos a separar las Series y las peliculas


def separar_titulos_por_tipo(df):
    df['Movies_title'] = np.where(df['type'] == 'Movie', df['title'], None)
    df['TVshow_title'] = np.where(df['type'] == 'TV Show', df['title'], None)
    return df

Netflix = separar_titulos_por_tipo(Netflix)

# Imprimir las columnas resultantes

print(Netflix[['title', 'Movies_title', 'TVshow_title']].head(10))
```

```python

# Tomamos el primer valor de la columna listed_in

def obtener_primer_dato(df, columna):

    return df[columna].str.split(',').str.get(0)

Netflix['One_gender'] = obtener_primer_dato(Netflix, 'listed_in')

# Verificar el resultado imprimiendo

print(Netflix[['listed_in', 'One_gender']].head(10))
```
